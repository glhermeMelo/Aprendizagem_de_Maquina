```python
#Array de uma dimensão
one_dimensional_array = np.array([1.2, 2.4, 3.5, 4.7, 6.1, 7.2, 8.3, 9.5])
one_dimensional_array
#print(one_dimensional_array)
```

```python
#Preencher arrau com zeros e uns
array_zeros = np.zeros(5)
print(array_zeros)

array_ones = np.ones([2,4])
print(array_ones)
```

```python
#Preencher array com uma sequência de números[limite_inferior, limite_superior]
#O intervalo é fechado no limite inferior e aberto no superior
sequence_of_integers = np.arange(5,12)
print(sequence_of_integers)
```

```python
#Preencher array com números inteiros aleatórios
random_integers_between_50_and_100 = np.random.randint(low=50, high=101, size=(6))
print(random_integers_between_50_and_100)
```

```python
#Preencher array com números float aleatórios
random_floats_between_0_and_1 = np.random.random([6])
print(random_floats_between_0_and_1)
```

# Tarefa: Criar um dataset linear
* Criar um array "feature" com uma sequência de inteiros de 6 a 20 (inclusive)
* Criar um array "label" onde :
label = (3)(feature) + 4

```python
feature = np.arange(6,21)
print(feature)

label = (feature * 3) + 4
print(label)
```
